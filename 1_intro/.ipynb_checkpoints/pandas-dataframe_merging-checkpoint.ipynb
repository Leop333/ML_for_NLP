# Предобработка данных


задача - вытащить данные из двух датасетов, проставить лейблы (fake = 0, true = 1), смерджить в один датасет, зашаффлить строки


```python
# ячейка импортов

import numpy as np
import pandas as pd
```

```python
# читаем данные

data_fake = pd.read_csv('/Users/macbook/Documents/datasets/news/Fake.csv',
                        encoding='utf-8')

# добавим колонку -- тип новости (true = 1, fake = 0)
data_fake['label'] = 0
data_fake.head(3)
```

```python
# теперь для тру-новостей

data_true = pd.read_csv('/Users/macbook/Documents/datasets/news/True.csv',
                        encoding='utf-8')

data_true['label'] = 1 # лейбл тру-новостей
data_true.head(3)
```

```python
# мерджим датафреймы
data = data_fake.append(data_true)
data
```

```python
# шаффлим строки

from sklearn.utils import shuffle
data = shuffle(data).reset_index(drop=True)
data
```

```python
# сохраняем в csv
data.to_csv("news_fake-n-true.csv")

```
