# Обзор

## что такое нейросеть


Нейронные сети - это тип алгоритма машинного обучения. Они(сети) состоят из слоев взаимосвязанных обрабатывающих узлов(также известных как нейроны), которые работают вместе для обработки входных данных и  выполнения задачи, поставленной перед алгоритмом: прогнозирования, классификации,  сегментации, генерации и тд.

Нейронные сети стали все более популярными в последние годы ([хотя были с нами с 40х-50х](https://en.wikipedia.org/wiki/Timeline_of_machine_learning)) благодаря их способности учиться на больших объемах данных и делать сложные прогнозы или решения на основе этих данных. Они показали большой потенциал в обработке естественного языка (NLP), где они были использованы для задач, таких как анализ тональности, перевод языка, распознавание речи и генерация текста.

В контексте NLP нейронные сети были особенно эффективны в захвате нюансов и тонкостей человеческого языка. Моделируя отношения между словами и фразами, они могут генерировать более точные и естественно звучащие ответы на входной текст.

Еще одним преимуществом нейронных сетей является их способность обрабатывать сложные и шумные данные.
Порой попадаются данные, при работе с которыми классические алгоритмы машинного обучения не приносят результатов. Самое время обратить взгляд на нейросети!

## что делать с нейросетью


- определяемся с данными, подготавливаем их (data pre-rocessing)
- определяемся с архитектурой сетки (+ строим ее)
- сеть обучается (feed forward, back propagation, loss)
- оцениваем результаты (evaluation)
- повторяем предыдущие два пункта при необходимости
- сохраняем обученную модель 
- используем по назначению (deployment)
- дообучаем, при необходимости (CI, CL, CD)


**как происходит обучение:** <br>
Когда мы говорим, что нейросеть "учится", мы подразумеваем процесс *регулирования силы связей между нейронами* c целью уменьшения значения функции ошибки. Нейросеть можно рассматривать как [сложную функцию](https://studymachinelearning.com/mathematics-behind-the-neural-network/#:~:text=Neural%20Network%20is%20a%20sophisticated,variable%20and%20learn%20the%20patterns.) (или набор связанных функций).

задача сети при обучении -- правильно расставить веса между слоями нейронов. В этом поможет оптимизация функции ошибки (loss function)

Сначала все веса просто расставлены случайно *(это называется инициализация весов)*, мы показываем сети данные, она выдаёт некоторый случайный ответ.<br>
Затем мы сравниваем, насколько результат отличается от нужного нам *(считаем значение функции ошибки)*. Затем идём по сети в обратном направлении, от выходного слоя к его предшествующему, от того - к его предшествуннику и так до входного слоя.  На каждом из этих шагов  проверяем, как сработал каждый нейрон: какие связи привели к слишком большому значению нейрона(их стоит погасить), а какие наоборот стоит усилить.<br>
Спустя большое количество таких циклов связи между нейронами в сети корректируются в сторону правильных. 

В больших сетях все это делается не вручную, а функциями.<br>

# Подробнее

## что такое нейрон


- Нейрон -- это абстракция. Нейрон можно представлять как функцию с кучей входов и одним выходом. Обычно результат работы этой функции существует на промежутке $[0,1]$<br> 
- Задача нейрона — взять числа от всех входов, выполнить над ними функцию и сохранить результат (это называется активацией нейрона).<br> 
- *Пример нейрона:* просуммировать все цифры со входов, и если их сумма больше N — выдать на выход единицу, иначе — ноль.

- Слои нейросети состоят из нейронов со значениями их активации
- на первом слое кол-во нейронов = кол-во признаков(features) у сэмпла данных из датасета

## связи и bias


- Связи — это каналы, которыми соединены нейроны предыдущего и следующего слоя. У каждой связи есть свой вес (можно представить как важность связи).
  - *Пример:* когда через связь с весом 0.5 проходит число 10, оно превращается в 5.
  
У связей есть веса. Веса в связах нужны, чтобы управлять на какие входы нейрон должен реагировать, а на какие нет.
- bias - дополнительный параметр, который корректирует значения финальной взвешенной суммы нейрона (он показывает, насколько значение финальной взвешенной суммы необходимо поднять/опустить, перед тем как нейрон активируется)

## активация

- активация -- это некоторое значение нейрона во время обучения, результат выполнения функции, заданной нейроном 

- во время обучения активация нейронов одного слоя влияет на активацию нейронов следующего
- мы не можем напрямую задавать значения активаций, но можем влиять на веса связей между слоями
- активации бывают [разными](http://rasbt.github.io/mlxtend/user_guide/general_concepts/activation-functions/)
- почти все функции активации нелинейны. Это специально: нелинейность нужна, чтобы нейросеть в процессе обучения смогла провести нелинейную границу между классами обучающих данных *(именно позволяет нейросетям решать задачи с нелинейной зависимостью)*

## как записать слои и веса в матричном виде: 

**forward feed** 



- нейроны одного слоя (с их значениями) -- как вектор-столбец $x$ 
- веса -- как матрица $W$ , где строка -- это веса связей от нейронов одного слоя к конкретному нейрону следующего слоя
- не забываем про байесы --  $b$
- тогда финальная взвешенная сумма = $Wa+b$
- при перемножении $Wa$ получается новый вектор-столбец. Значения в нем --  это активации нейронов в новом слое 
    - если хотим, чтобы слой был нелинейным, обернем все проиведение в нелинейную функцию активации: $Out = f(Wa+b)$
- один слой посчитан!
    - следующие считаем по такому же принципу

ок, вот так обучаемся. А как узнать, насколько мы ошиблись? 

## лосс

- Loss Function это функция-показатель, насколько сеть ошиблась в предсказании *(эталон vs предсказанное значение)*.
- при обучении модели, функция ошибки на вход принимает результат работы всех весов и байесов модели (которые выдали предсказание) и выдает число, описывающее, насколько велика ошибка
- основная задача обучения сети (как и в классическом ML) -- минимизировать значение функции ошибки

## обратное распространение ошибки (backpropagation)

backpropagation -- это метод посчитать градиент 

- градиент-вектор -- это вектор из компонентов по числу всех связей и баесов в сети, на всех слоях. Думать о нем в пространстве не очень удобно, но можно представить удобнее: некоторые компоненты (с большими позитивными или негативными значениями) -- это очень важные для обучения модели(т.е. для лосс-функции) параметры (связи между некоторыми нейронами или баесы). 

<img src="https://i.stack.imgur.com/xwlAI.png"> 
- чтобы минимизировать функцию ошибки, нужно повлиять на активацию нейронов последнего слоя. Чтобы повлиять на эту активацию, нужно изменить веса предыдущего слоя. И так далее, до самого начала.  Понять, в какую сторону их менять, помогает градиент лосс-функции. 
(нужно двигаться в противоположном направлении)




**чуть больше математики:**

## Feed-Forward NNs

Прямые нейронные сети - это тип нейронной сети, где информация течет только в одном направлении: от входного слоя к выходному слою. Эти сети состоят из нескольких слоев взаимосвязанных нейронов, при этом каждый слой трансформирует входные данные нелинейным образом.

Базовая структура прямой нейронной сети состоит из трех типов слоев: 
- входного слоя
- скрытых слоев
- выходного слоя. 
Входной слой получает входные данные, которые затем передаются через скрытые слои, где данные трансформируются с помощью набора весов и смещений, связанных с каждым нейроном. Выходной слой производит окончательный прогноз или классификацию на основе трансформированных данных.



Для обучения прямой нейронной сети мы используем алгоритм *обратного распространения ошибки*. Этот алгоритм работает путем вычисления ошибки между прогнозируемым выходом и истинным выходом для заданного входа, а затем распространяет эту ошибку назад через сеть, чтобы обновить веса и смещения, связанные с каждым нейроном. Повторяя этот процесс на большом наборе данных, сеть постепенно учится делать более точные предсказания.

Прямые нейронные сети применяются в широком спектре задач NLP, включая анализ тональности, распознавание именованных сущностей и классификацию текста. Например, прямая нейронная сеть может быть обучена на датасете обзоров на фильмы, чтобы предсказывать эмоциональную окраску (положительную или отрицательную) некоторого обзора.

## RNNs



Рекуррентные нейронные сети - это тип нейронных сетей, которые могут работать с последовательными данными, такими как тексты, речь или временные ряды. Эти сети имеют внутреннюю память, которая позволяет им сохранять информацию о предыдущих входах и использовать ее при обработке последующих входов.

Базовая структура рекуррентной нейронной сети состоит из трех типов слоев: 
- входного слоя 
- скрытых слоев
- выходного слоя. 

Однако в отличие от прямых нейронных сетей, входные данные подаются на каждый временной шаг вместе с информацией о предыдущем состоянии сети. Таким образом, рекуррентная нейронная сеть может работать с последовательными данными, обрабатывая их по одному шагу за раз и сохраняя информацию о предыдущих шагах.

Самый простой тип рекуррентной нейронной сети - это LSTM. LSTMs были разработаны для решения проблемы исчезающего градиента, которая возникает при обучении рекуррентных нейронных сетей на длинных последовательностях данных. Они имеют внутренний механизм, который позволяет им сохранять информацию о предыдущих состояниях сети на более длительное время.


Рекуррентные нейронные сети были применены в широком спектре задач NLP, включая языковой перевод, распознавание речи и генерацию текста. Например, рекуррентная нейронная сеть может быть обучена на парах предложений на двух языках, чтобы переводить один язык на другой.

Одной из особенностей рекуррентных нейронных сетей является их способность к моделированию контекста. Поскольку они сохраняют информацию о предыдущих состояниях сети, они могут учитывать контекст при обработке последующих входов. Это может привести к более точным и осмысленным ответам на задачи NLP.

Однако рекуррентные нейронные сети также имеют свои ограничения. Одной из проблем является склонность к забыванию дальних зависимостей в последовательных данных. Чем дальше во времени расположены данные, тем сложнее для рекуррентной нейронной сети сохранить информацию о них.

## CNNs



Сверточные нейронные сети - это тип нейронных сетей, которые эффективно работают с данными, имеющими пространственную структуру, такими как изображения или тексты. Эти сети используют операцию свертки для обнаружения локальных паттернов во входных данных и выделения более высокоуровневых признаков.

Базовая структура сверточной нейронной сети состоит из трех типов слоев: сверточного слоя, субдискретизирующего слоя и полносвязного слоя. Сверточный слой применяет несколько фильтров (ядер) к входным данным, чтобы найти паттерны в определенных областях данных. Субдискретизирующий слой уменьшает размерность данных, сохраняя важную информацию. Полносвязный слой преобразует признаки, выделенные в предыдущих слоях, в конечный результат.

Сверточные нейронные сети широко применяются в обработке изображений, но они также могут быть использованы для обработки текстовых данных, таких как наборы слов или символов. Например, сверточная нейронная сеть может быть обучена на наборе текстовых документов для классификации документов по тематикам.

Одной из особенностей сверточных нейронных сетей является их способность к автоматическому извлечению признаков из входных данных. Это позволяет им выделять абстрактные признаки, такие как края или текстуры, без необходимости явно задавать эти признаки.

Однако сверточные нейронные сети также могут столкнуться с проблемой переобучения, особенно если количество тренировочных данных недостаточно большое. Для решения этой проблемы могут быть использованы различные методы, такие как регуляризация и аугментация данных.

## Seq2Seq and Transformers



Seq2Seq и Трансформеры - это два из самых популярных типов нейронных сетей, которые применяются в обработке естественного языка.

Seq2Seq (Sequence-to-Sequence) - это архитектура нейронных сетей, которая была разработана для обработки последовательностей разной длины и имеет две составляющие: энкодер и декодер. Энкодер принимает на вход последовательность данных и преобразует ее в фиксированный вектор, который затем передается в декодер, который генерирует новую последовательность данных на основе входного вектора. Seq2Seq может быть использована для решения различных задач, таких как машинный перевод, генерация текста и чат-боты.

Однако, Seq2Seq имеет свои ограничения, такие как недостаточную способность к моделированию длинных зависимостей в данных. Эта проблема может быть решена с помощью Трансформеров.

Трансформеры - это архитектура нейронных сетей, которая была представлена в 2017 году. Трансформеры используют внимание (attention) вместо рекуррентных или сверточных слоев для моделирования зависимостей между входными и выходными данными. Трансформеры могут работать с последовательностями разной длины и могут обрабатывать больше информации за один проход по сравнению с рекуррентными и сверточными нейронными сетями.

Трансформеры были использованы для решения многих задач NLP, таких как машинный перевод, генерация текста, ответы на вопросы и распознавание именованных сущностей. Они демонстрируют высокую точность и позволяют обрабатывать длинные последовательности данных более эффективно.

Однако, Трансформеры могут быть более сложными для обучения и требуют большого количества вычислительных ресурсов. Некоторые модификации, такие как BERT (Bidirectional Encoder Representations from Transformers), были разработаны для более эффективного использования Трансформеров и достижения лучших результатов на различных задачах NLP.

## Attention

Внимание (Attention) - это механизм, который позволяет нейронным сетям фокусироваться на наиболее важных частях входных данных при выполнении задач обработки естественного языка. Внимание может быть использовано в различных типах нейронных сетей, таких как Рекуррентные и Трансформеры.

Механизм внимания позволяет нейронной сети определить, какие части входных данных важнее для выполнения задачи, и сосредоточиться на них. Например, при переводе фразы с одного языка на другой, нейронная сеть может использовать механизм внимания, чтобы определить, какие слова в исходной фразе наиболее важны для перевода, и использовать эту информацию при генерации перевода.

Внимание может быть реализовано в различных формах, таких как множественное внимание (Multi-Head Attention), который позволяет нейронной сети фокусироваться на нескольких аспектах входных данных одновременно, или само-внимание (Self-Attention), который позволяет нейронной сети определить, какие части ее собственного вывода наиболее важны для выполнения задачи.

Механизм внимания может значительно повысить точность и эффективность работы нейронных сетей в обработке естественного языка. Он может использоваться для выполнения различных задач, таких как машинный перевод, ответы на вопросы и генерация текста.

Однако, внимание также может быть сложным для реализации и обучения, особенно если используется в сочетании с другими сложными типами нейронных сетей, такими как Трансформеры. Требуется большое количество вычислительных ресурсов и опыта в обработке данных для достижения хороших результатов с использованием механизма внимания.
